# Fetching Data From an API

## TMDB movie website take top rank movies with the help of API

```python
import pandas as pd
```

```python
import requests
```

## take data with the help of URL

```python
response=requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1' )
```

## print only Results table

```python
response.json()['results']
```

```python
pd.DataFrame(response.json()['results']).head()
```

## now this data convert to dataframe

```python
pd.DataFrame(response.json()['results']).head()[['id','title','overview','release_date','vote_average','vote_count']]
```

```python
df=pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','vote_average','vote_count']]
```

# same program in function using loop

```python
df=pd.DataFrame()
```

```python
for i in range (1,429):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','vote_average','vote_count']]
    df = pd.concat([df,temp_df],ignore_index=True)
```
