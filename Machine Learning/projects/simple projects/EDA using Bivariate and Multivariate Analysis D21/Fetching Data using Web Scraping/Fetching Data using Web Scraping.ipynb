# Fetching Data using Web Scraping

### we open a car websit where car is selling so we take some infomation like car name , price , model , used in KM , fuel type ,  CC and car type  

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 "
                  "(KHTML, like Gecko) Chrome/124.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9"
}
webpage=requests.get('https://www.pakwheels.com/used-cars/peshawar/24821',headers=headers).text
```

now your request is working 
# now convert to HTML format

```python
soup=BeautifulSoup(webpage,'lxml')
```

### now i am taking car names 

```python
car_names=[]
car_n=soup.find_all('h3',style='white-space: normal;')

for i in car_n:
    car_names.append(i.text.strip())
```

### car price


```python
car_price=[]

car_p=soup.find_all('div',class_='price-details generic-dark-grey')

for j in car_p:
    car_price.append(j.text.strip())
```

#### Now  model , used , fuel type ,  cc and car type take in one loop beacuse they all in one same class so we access with index

```python
car_model=[]
car_used=[]
car_fuel_type=[]
cc=[]
car_type=[]

car_detail=soup.find_all('ul', class_='list-unstyled search-vehicle-info-2 fs13')

for k in car_detail:
    car_model.append(k.find_all('li')[0].text.strip())
    car_used.append(k.find_all('li')[1].text.strip())
    car_fuel_type.append(k.find_all('li')[2].text.strip())
    cc.append(k.find_all('li')[3].text.strip())
    car_type.append(k.find_all('li')[4].text.strip())
    
```

```python
d={'Car Name':car_names,'Car Price':car_price,'Car Model':car_model,'Mileage':car_used,'Fuel Type':car_fuel_type,'Engine Capacity':cc,'Car Type':car_type}
```

## data convert to Data Frame 

```python
df=pd.DataFrame(d)
```
